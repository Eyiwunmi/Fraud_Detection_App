```python
# ============================================
# 0. Imports
# ============================================
import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from catboost import CatBoostClassifier
import joblib

# ============================================
# 1. Feature Engineering Transformer
# ============================================
class FeatureEngineer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        df = X.copy()
        
        # Convert datetime
        df['transaction_time'] = pd.to_datetime(df['trans_date_trans_time'])
        df['dob'] = pd.to_datetime(df['dob'])
        
        # Extract datetime features
        df['transaction_hour'] = df['transaction_time'].dt.hour
        df['transaction_day'] = df['transaction_time'].dt.day
        df['transaction_month'] = df['transaction_time'].dt.month
        
        # Age
        df['age'] = (df['transaction_time'] - df['dob']).dt.days // 365
        
        # Log of amount
        df['amt_log'] = np.log1p(df['amt'])
        
        # Haversine distance
        def haversine(lat1, lon1, lat2, lon2):
            R = 6371
            lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
            dlat = lat2 - lat1
            dlon = lon2 - lon1
            a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
            c = 2 * np.arcsin(np.sqrt(a))
            return R * c
        
        df['distance'] = haversine(df['lat'], df['long'], df['merch_lat'], df['merch_long'])
        
        # Features to use
        feature_cols = [
            'amt', 'amt_log', 'transaction_hour', 'transaction_day', 'transaction_month',
            'age', 'distance', 'lat', 'long', 'merch_lat', 'merch_long', 'city_pop'
        ]
        
        return df[feature_cols]

# ============================================
# 2. Numeric Scaler
# ============================================
scaler = StandardScaler()

# ============================================
# 3. CatBoost Classifier
# ============================================
catboost_model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05,
    depth=8,
    loss_function='Logloss',
    eval_metric='AUC',
    class_weights=[1, 40],  # upweight fraud class
    verbose=100
)

# ============================================
# 4. Full Pipeline
# ============================================
pipeline = Pipeline([
    ('feature_engineer', FeatureEngineer()),
    ('scaler', scaler),
    ('catboost', catboost_model)
])

# ============================================
# 5. Train/Test Split
# ============================================
from sklearn.model_selection import train_test_split

# df_train should contain raw training data
X = df_train  # raw dataframe
y = df_train['is_fraud']

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# ============================================
# 6. Fit Pipeline
# ============================================
pipeline.fit(X_train, y_train)

# ============================================
# 7. Evaluate on Validation
# ============================================
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

y_val_proba = pipeline.predict_proba(X_val)[:, 1]
THRESHOLD = 0.7  # adjust to control precision/recall
y_val_pred = (y_val_proba >= THRESHOLD).astype(int)

print("Classification Report (Validation):\n", classification_report(y_val, y_val_pred))
print("ROC-AUC (Validation):", roc_auc_score(y_val, y_val_proba))
print("Confusion Matrix (Validation):\n", confusion_matrix(y_val, y_val_pred))

# ============================================
# 8. Save Pipeline for Later Use
# ============================================
joblib.dump(pipeline, 'fraud_detection_pipeline.pkl')
print("Pipeline saved as 'fraud_detection_pipeline.pkl'")

# ============================================
# 9. Predict on Raw Test Set
# ============================================
# df_test should be raw test dataframe
y_test_proba = pipeline.predict_proba(df_test)[:, 1]
y_test_pred = (y_test_proba >= THRESHOLD).astype(int)

print("Classification Report (Test):\n", classification_report(df_test['is_fraud'], y_test_pred))
print("ROC-AUC (Test):", roc_auc_score(df_test['is_fraud'], y_test_proba))
print("Confusion Matrix (Test):\n", confusion_matrix(df_test['is_fraud'], y_test_pred))
```
